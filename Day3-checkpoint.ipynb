```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:/Users/Aatif/Downloads/Onion data.csv"
onion_data = pd.read_csv(file_path)

```

```python
onion_data.columns = onion_data.columns.str.strip()

# Convert the 'Price Date' column to datetime format
onion_data['Price Date'] = pd.to_datetime(onion_data['Price Date'], format='%d/%m/%Y')

# Filter data between 1 Jan 2025 and 8 Apr 2025
onion_data_filtered = onion_data[(onion_data['Price Date'] >= '2025-01-01') & (onion_data['Price Date'] <= '2025-04-08')]

# Sort the data by date
onion_data_sorted = onion_data_filtered.sort_values(by='Price Date')

# Plotting the Minimum, Maximum, and Modal prices
plt.figure(figsize=(10, 6))
plt.plot(onion_data_sorted['Price Date'], onion_data_sorted['Min Price\n(Rs./Quintal)'], label='Min Price', color='blue')
plt.plot(onion_data_sorted['Price Date'], onion_data_sorted['Max Price\n(Rs./Quintal)'], label='Max Price', color='red')
plt.plot(onion_data_sorted['Price Date'], onion_data_sorted['Modal Price\n(Rs./Quintal)'], label='Modal Price', color='green')

# Formatting the plot
plt.title('Onion Price Trends (1 Jan 2025 - 8 Apr 2025)')
plt.xlabel('Date')
plt.ylabel('Price (Rs./Quintal)')
plt.legend(loc='best')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
